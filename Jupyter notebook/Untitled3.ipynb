```python
import pandas as pd
import random
from faker import Faker

fake = Faker()

# Load patients so visits can link correctly
patients_df = pd.read_csv(
    r"D:\python project files\Health Care Revenue Cycle & Patient Risk Analytics\data\raw\patients.csv"
)

# Number of visits
num_visits = 200000

# Mapping chronic condition → department + diagnosis
condition_map = {
    "Diabetes": {"department": "General Medicine", "diagnosis": "Diabetes"},
    "Hypertension": {"department": "General Medicine", "diagnosis": "Hypertension"},
    "Heart Disease": {"department": "Cardiology", "diagnosis": "Chest Pain"},
    "Asthma": {"department": "Pulmonology", "diagnosis": "Asthma"},  # you can add Pulmonology
    "None": {"department": "General Medicine", "diagnosis": "Fever"}  # fallback
}

payment_status_list = ["Paid", "Pending", "Denied"]

visits = []

for i in range(1, num_visits + 1):
    # Pick a random patient record
    patient = random.choice(patients_df.to_dict('records'))
    chronic_condition = patient['chronic_condition']

    # 70% chance to follow mapping, 30% random for variety
    if chronic_condition in condition_map and random.random() < 0.7:
        dept = condition_map[chronic_condition]['department']
        diag = condition_map[chronic_condition]['diagnosis']
    else:
        # fallback random choice
        dept = random.choice([
            "Cardiology", "Orthopedics", "Neurology", "Emergency",
            "Pediatrics", "Oncology", "General Medicine"
        ])
        diag = random.choice([
            "Diabetes", "Hypertension", "Fracture", "Chest Pain",
            "Migraine", "Asthma", "Fever", "Cancer"
        ])

    visit = {
        "visit_id": i,
        "patient_id": patient['patient_id'],
        "visit_date": fake.date_between(start_date='-2y', end_date='today'),
        "department": dept,
        "doctor_id": random.randint(1001, 1100),
        "diagnosis": diag,
        "treatment_cost": round(random.uniform(100, 15000), 2),
        "payment_status": random.choice(payment_status_list)
    }

    visits.append(visit)

# Create dataframe
visits_df = pd.DataFrame(visits)

# Save csv
visits_df.to_csv(
    r"D:\python project files\Health Care Revenue Cycle & Patient Risk Analytics\data\raw\visits.csv",
    index=False
)

print("visits.csv created successfully")
print(visits_df.head())
```
