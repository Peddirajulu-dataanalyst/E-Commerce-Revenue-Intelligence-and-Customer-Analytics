```python
#load data
import pandas as pd
orders = pd.read_csv(r"D:\python project files\E-commerce Revenue Intelligence and Customer Analytics\data\raw\orders.csv")
customers = pd.read_csv(r"D:\python project files\E-commerce Revenue Intelligence and Customer Analytics\data\raw\customers.csv")
products = pd.read_csv(r"D:\python project files\E-commerce Revenue Intelligence and Customer Analytics\data\raw\products.csv")
returns = pd.read_csv(r"D:\python project files\E-commerce Revenue Intelligence and Customer Analytics\data\raw\returns.csv")
```

```python
#order month / year (time Analysis)
import pandas as pd 
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['order_month'] = orders['order_date'].dt.to_period('M')
orders['order_year'] = orders['order_date'].dt.year
```

```python
#profit margin
orders['profit_margin'] = orders['profit'] / orders['sales']
```

```python
#discount bucket
orders['discount_bucket'] = pd.cut(
    orders['discount'],
    bins=[0,0.1,0.2,0.3,1],
    labels=['Low','Medium','High','Very High']
)
```

```python
#customer revenue
customer_revenue = orders.groupby('customer_id')['sales'].sum()
orders['customer_revenue'] = orders['customer_id'].map(customer_revenue)
```

```python
#order frequency
order_freq = orders.groupby('customer_id').size()
orders['order_frequency'] = orders['customer_id'].map(order_freq)
```

```python
#Recency
last_date = orders['order_date'].max()

recency = orders.groupby('customer_id')['order_date'].max()
recency = (last_date - recency).dt.days

orders['recency'] = orders['customer_id'].map(recency)
```

```python
#rfm score 
rfm = orders.groupby('customer_id').agg({
    'order_date':'max',
    'order_id':'count',
    'sales':'sum'
})
```
